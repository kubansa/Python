```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```
# generowanie danych
# generuje dane od 0 do 20 równo rozłozone 100 liczb
x = np.linspace(0,20,100)
# rozkład normalny 4 próbki (4 zmienne, po 100 liczb)
y = np.random.randn(100,4)
# suma skumulowana, po kolumnach inaczej wzgledm osi = 0
y = np.cumsum(y,axis = 0)
```

```
plt.plot(x,y)
```

```
# ustawienei motywu
sns.set()
plt.plot(x,y)
```

## wykres relacyjny

```
# wykres relacyjny
sns.relplot(data = df, x = 'total_bill', y = 'tip')
```

```
# ustawienei stylu
sns.set(style = 'whitegrid', palette = 'pastel', font_scale=1)
sns.relplot(data = df, x = 'total_bill', y = 'tip', hue = 'sex', size = 'sex', sizes=(10,200), palette = 'viridis', style = 'sex')
```

```
sns.relplot(data = df, x = 'total_bill', y = 'tip', hue = 'size', size = 'size', sizes=(10,200), palette = 'viridis', col = 'time')
```

```
sns.relplot(data = df, x = 'total_bill', y = 'tip', hue = 'size', size = 'size', sizes=(10,200), palette = 'viridis', row = 'time')
```

```
sns.relplot(data = df, x = 'total_bill', y = 'tip', hue = 'size', size = 'size', sizes=(10,200), palette = 'viridis', col = 'time', row = 'smoker')
```

## wykres kategoryczny

```
df = sns.load_dataset('tips')
df.head()
```

```
sns.set(style = 'ticks')
sns.catplot(data = df, x = 'day',y='total_bill')
```

```
# ulepszenie wykresu za pomocą metody swarm
sns.catplot(data = df, x = 'day',y='total_bill', kind = 'swarm')
```

```
# wykres pudełkowy
sns.catplot(data = df, x = 'day',y='total_bill', kind = 'box')
```

```
# dla danych lategorycznych policzy ile wystepuje
sns.catplot(data = df, x = 'day', kind = 'count')
```

```
sns.catplot(data = df, x = 'day',y='total_bill', kind = 'swarm', hue = 'sex')
```

```
# wyciecie pewnych danych
sns.catplot(data = df.query("day == ['Fri','Sat','Sun']"), x = 'day',y='total_bill', kind = 'swarm', hue = 'sex')
```

## Wykres rozkładów (rozrzutu)

```
sns.set(style='darkgrid')
df = np.random.multivariate_normal(mean = [0,1], cov=[(1,-0.4), (-0.4,1)], size = 200)
df = pd.DataFrame(df, columns = ['x','y'])
df.head(3)
```

```
df.describe()
```

```
sns.jointplot(data = df, x='x', y= 'y')
```

```
sns.set(style = 'white')
sns.jointplot(data = df, x='x', y= 'y', kind = 'hex')
```

```
g= sns.jointplot(data = df, x='x', y= 'y', kind = 'kde', nlevel = 15, cmap = 'viridis')
g.plot_joint(plt.scatter, c= 'w', s= 30, marker = '+')
```
